```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
#histograms
plt.hist(df1['SAT'],color='orange')
plt.xlabel('SATSCORES')
plt.ylabel('Frequency')
plt.title('Histogram of SAT scores')
plt.show()
```

```python
sns.histplot(df1['SAT'],color='blue')#sns as seaborn
```

```python
plt.hist(df1['Accept'])
plt.xlabel('Acceptance Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Acceptance Rates')
plt.show()
```

```python
#sctter plots
plt.plot(df1['SAT'],df1['Accept'],'bo')#bo stands for blue color circles
plt.xlabel('SAT Scores')
plt.ylabel('Acceptance Rates')
plt.title('Scatter Plot of SAT Scores vs Acceptance Rates')
plt.show()
```

```python
#can you create a scatter plot using matplotlib between sat columns and 'top10' columns
plt.plot(df1['SAT'], df1['Top10'], 'o')
plt.xlabel('SAT Scores')
plt.ylabel('Top10 %')
plt.title('Scatter Plot of SAT Scores vs Top10 %')
plt.show()
```

```python
#pair plots
sns.pairplot(df1)
```
